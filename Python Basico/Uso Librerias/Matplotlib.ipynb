```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#los graficos basicos de funciones estan en numpy funciones 
```

```python
x = np.linspace(0,50)
y = x**2
plt.plot(x,y,'yx')
```

```python
#el subplot permite varias graficas de un dataframe
plt.subplot(1,2,1)#n filas, n columnas,index
plt.plot(x,y)#asocia grafico a la 1 fila
plt.subplot(1,2,2)
plt.hist(y)
```

```python
#grafico dentro de grafico XD
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.5,0.9])#dos primeros, posicion en lienzo, dos ultimos tamao 
ax2 = fig.add_axes([0.2,0.2,0.3,0.3])
ax.set_facecolor('yellow')
ax.plot(x,y)
ax2.plot(x,y*2)
```

```python
x = np.linspace(0,100,1000)
y = np.sin(x)*x**2
fig,axes = plt.subplots(1,2,figsize=(10,5))#fila,columna,
axes[0].plot(x,y,color = 'green', linestyle = '-',label = '$sin(x)$')#color, estilo linea, leyenda
axes[0].set_title('Relacion  X Y')
axes[0].set_xlabel("x")
axes[0].set_ylabel('sen(x)')
axes[1].plot(x,y**5,'yx')#1 color, 2 forma, tambien se puede en rgb
axes[1].set_title('Relacion x y**5')
axes[0].legend()
plt.style.use('fast')
```
