```python
import seaborn as sns #seaborn genera graficos a nivel mas general y util, esta basado en matplot
import matplotlib.pyplot as plt
import pandas as pd
tips = sns.load_dataset('tips')
print(tips)
```

```python
#set sns.set(stlye =, pplatte) configura estilos, permite metodos de plt
#displot es para todos los de distribucion
sns.displot(data = tips, x = 'tip', hue = 'sex', bins = 20, cumulative = False, multiple = 'dodge') # sin designacion da un histograma, stat es leyenda, multiple = layer, es default, tambien sirve stack
sns.displot(data = tips, x = 'tip', hue = 'sex', kind = 'kde') #grafico de densidad , cumulative lo da como F abs
sns.displot(data = tips, x = 'tip', hue = 'sex', kind = 'ecdf' ) #escalonado
```

```python
#categoricos, para texto
sns.catplot ( data = tips, x = 'day', y = 'total_bill', hue = 'sex', dodge = True, kind = 'strip') # por puntos, se ve la agrupacion por dia y tiempo, el de puntos es el default
sns.catplot ( data = tips, x = 'day', y = 'total_bill', hue = 'sex', dodge = True, kind = 'box') #un caja y bigotes XD
sns.catplot ( data = tips, x = 'day', y = 'total_bill', hue = 'sex', dodge = True, kind = 'swarm') # un strip que añade horizontal, para ver concentracion
sns.catplot ( data = tips, x = 'day', y = 'total_bill', hue = 'sex', dodge = True, kind = 'violin') # una distribucion de violin, osea una dsitribucion de densidad simetrica
```

```python
#por relacion, valores numericos con numericos
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', style = 'time', size = 'size', kind = 'scatter') #un scatter por default
plt.legend(loc = 'center', bbox_to_anchor = (1.12,0.6))
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', style = 'time', size = 'size', kind = 'line') # por lineas
```

```python
#joinplot, une graficos, genera 2
#pairplot, usa solo las numericos, las cruza entre si
sns.jointplot(data = tips, x = 'total_bill', y = 'tip',hue = 'sex') #ya depende del kind, da igual
sns.pairplot(data = tips)
```

```python
#heatmap
tips.corr() #tabla de correlaciones de valores numericos, obviamente la correlacion entre un mismo valor es 1 XD, ni que fuera pendejo
sns.heatmap(tips.corr(),annot = True, cmap = 'coolwarm')#grafica un mapa de color de acuerdo a los valroes de una tabla
```
