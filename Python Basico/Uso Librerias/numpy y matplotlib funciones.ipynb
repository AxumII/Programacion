```python
import matplotlib.pyplot as plt
import numpy as np
#con la 1 graficamos funciones, la 2 da mas matematica
```

```python
N = 100
m = 10
b = 5
def f(x):
    return m*x+b

x = np.linspace(-30,30,num = N)
y = f(x)

#print(y)
fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid
```

```python
#Polinomicas
N = 100
a = 5
b = 2
c = 10
def f(x):
    return (a*(x**2))+(b*x)+c

y = f(x)

plt.plot(x,y,'rx')#3 parametro, color y forma

```
