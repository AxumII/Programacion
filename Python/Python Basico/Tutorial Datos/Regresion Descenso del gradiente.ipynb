```python
#Esto es para el descenso del gradiente
import numpy as np
import scipy as sc #funciones extra cientificas
import matplotlib.pyplot as plt
#requiere que la funcion a usar sea derivable, por ende, debe ser continua
```

```python
#th es el vector de parametros
#funcion a usar
func = lambda th: ((np.sin(((1/2)*th[0]**2)-((1/4)*th[1]**2)+(3)))*(np.cos((2*th[0])+(1)-(np.e**th[1]))))

res = 100
X = np.linspace(-2,2,res)
Y = np.linspace(-2,2,res)
Z = np.zeros((res,res))

for ix, x in enumerate(X):#permite crear una matriz de z
    for iy, y in enumerate(Y):
        Z[iy,ix] = func([x,y])
#plt.contour? #permite ver como se usa un metodo
plt.contour(X,Y,Z,10000)#permite ve rcurvas de nivel en 2d
plt.colorbar()

Tt = np.random.rand(2)*4-2
plt.plot(Tt[0],Tt[1],'x',c = 'red')

ttt = np.copy(Tt)
h = 0.001
grad = np.zeros(2)
lr = 0.05

for _ in range(1000):
    for it, th in enumerate(Tt):
        ttt = np.copy(Tt)
        ttt[it] = ttt[it] + h
        Dv = (func(ttt) - func(Tt))/h
        grad[it] = Dv

    Tt = Tt -lr* grad
    if(_%10==0):
        plt.plot(Tt[0],Tt[1],'x',c ='red')

plt.plot(Tt[0],Tt[1],'x',c = 'red')
```
