```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5,6,7,8])

y = np.array([1,4,8,14,30,54,102,458])
print(y)
plt.plot(x,y,"o")
plt.xlabel('Exponente de base 10')
plt.ylabel('Tiempo(ms)')
plt.title('Creacion libros iguales')
plt.show()
#0.68
```
